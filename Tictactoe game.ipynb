```python
import random

# Function to draw the Tic Tac Toe board
def draw_board(board):
    print(" " + board[1] + " | " + board[2] + " | " + board[3])
    print("-----------")
    print(" " + board[4] + " | " + board[5] + " | " + board[6])
    print("-----------")
    print(" " + board[7] + " | " + board[8] + " | " + board[9])

# Function to check if a player has won
def check_win(board, player):
    return (
        (board[1] == board[2] == board[3] == player) or
        (board[4] == board[5] == board[6] == player) or
        (board[7] == board[8] == board[9] == player) or
        (board[1] == board[4] == board[7] == player) or
        (board[2] == board[5] == board[8] == player) or
        (board[3] == board[6] == board[9] == player) or
        (board[1] == board[5] == board[9] == player) or
        (board[3] == board[5] == board[7] == player)
    )

# Function for the computer's turn
def computer_turn(board):
    # Check if there is an opportunity to win
    for i in range(1, 10):
        if board[i] == " ":
            board[i] = "O"
            if check_win(board, "O"):
                return i
            else:
                board[i] = " "

    # Check if the player has an opportunity to win and block it
    for i in range(1, 10):
        if board[i] == " ":
            board[i] = "X"
            if check_win(board, "X"):
                return i
            else:
                board[i] = " "

    # Choose a random empty spot
    empty_spots = [i for i in range(1, 10) if board[i] == " "]
    if empty_spots:
        return random.choice(empty_spots)

    return None

# Function to play the game
def play_game():
    board = [" "] * 10
    current_player = "X"

    while True:
        draw_board(board)

        if current_player == "X":
            move = int(input("Enter your move (1-9): "))
            if move < 1 or move > 9 or board[move] != " ":
                print("Invalid move! Try again.")
                continue
            board[move] = current_player
        else:
            print("Computer's turn...")
            move = computer_turn(board)
            if move is None:
                print("It's a draw!")
                break
            board[move] = current_player

        if check_win(board, current_player):
            draw_board(board)
            print("Congratulations! Player " + current_player + " wins!")
            break

        if " " not in board[1:]:
            draw_board(board)
            print("It's a draw!")
            break

        current_player = "O" if current_player == "X" else "X"

# Start the game
play_game()
```
